```
from pyspark.sql.functions import *
import urllib
ACCESS_KEY = aws_key.where(col('User name')=='karz').select('Access key ID').collect()[0]['Access key ID']
SECRET_KEY = aws_key.where(col('User name')=='karz').select('Secret access key').collect()[0]['Secret access key']
ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe="")
```

```
dbutils.fs.unmount("/mnt/karz-bigdata-su22")
AWS_S3_BUCKET = "karz-bigdata-su22"
MOUNT_NAME = "/mnt/karz-bigdata-su22"
SOURCE_URL = "s3n://{0}:{1}@{2}".format(ACCESS_KEY, ENCODED_SECRET_KEY, AWS_S3_BUCKET)
dbutils.fs.mount(SOURCE_URL, MOUNT_NAME)
```

```
input_file = sc.textFile('/FileStore/tables/a2q1_input.txt')
```

```
input_path = input_file.collect()
ipath = input_path[0]
iteration = int(input_path[1])
opath = input_path[2]
```

```
alpha=0.15
initialPR=10
```

```
#Using DF to remove header info
airportData=spark.read.option("header","true").option("inferSchema","true").csv(ipath)
```

```
display(airportData)
```

```
rddAirportData=airportData.rdd
```

```
rddAirportData.collect()
```

```
distinctAirport=rddAirportData.flatMap(lambda x: [x[0],x[2]]).distinct()
```

```
N=distinctAirport.count()
```

```
N
```

```
sourceNodes=distinctAirport.subtract(rddAirportData.map(lambda x: x[2]).distinct()).map(lambda x: (x,alpha/N))
```

```
sourceNodes.collect()
```

```
PR=distinctAirport.map(lambda x: [x,initialPR]).toDF(["AIRPORT","PR_RANK"])
```

```
 PR.count()
```

```
originAirport=rddAirportData.map(lambda x: x[0])
```

```
originAirport.collect()
```

```
outDegree=originAirport.map(lambda x: (x,1)).reduceByKey(lambda x, y: x + y)
```

```
outDegree.collect()
```

```
from pyspark.sql.functions import desc

for i in range(0, iteration):
    outDegreePR=PR.join(outDegree.toDF(["AIRPORT","OutDegree"]),"AIRPORT")
    outDegreePR=outDegreePR.rdd.map(lambda x: (x[0],x[1]/x[2]))
    dfEdgeAirport=rddAirportData.map(lambda x: [x[0],x[2]]).toDF(["AIRPORT","dest"])
    dfEdgeAirport=dfEdgeAirport.join(outDegreePR.toDF(["AIRPORT","PRData"]),["AIRPORT"])
    inDegreeAirport=dfEdgeAirport.rdd.map(lambda x: (x[1],x[2])).reduceByKey(lambda x,y: x+y)
    PR=sourceNodes.union(inDegreeAirport.map(lambda x: [x[0],(alpha*(1/N))+((1-alpha)*x[1])])).toDF(["AIRPORT","PR_RANK"])
```

```
output = PR.orderBy(desc("PR_RANK"))
display(output)
```

```
output.coalesce(1).write.format('csv').option('header','true').save(opath)
# The text file has been written to the AWS S3 bucket and it can be accessed by URI: "s3://karz-bigdata-su22/PageRankOutput.csv/"
```
