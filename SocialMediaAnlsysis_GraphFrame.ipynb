```
aws_key = spark.read.format("csv").option("header", "true").option("sep", ",").load("/FileStore/tables/new_user_credentials.csv")
```

```
from pyspark.sql.functions import *
import urllib
ACCESS_KEY = aws_key.where(col('User name')=='karzz').select('Access key ID').collect()[0]['Access key ID']
SECRET_KEY = aws_key.where(col('User name')=='karzz').select('Secret access key').collect()[0]['Secret access key']
ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe="")
```

```
# dbutils.fs.unmount("/mnt/karz-bigdata-su22")
AWS_S3_BUCKET = "karz-bigdata-su22"
MOUNT_NAME = "/mnt/karz-bigdata-su22"
SOURCE_URL = "s3n://{0}:{1}@{2}".format(ACCESS_KEY, ENCODED_SECRET_KEY, AWS_S3_BUCKET)
dbutils.fs.mount(SOURCE_URL, MOUNT_NAME)
```

```
# /FileStore/tables/assignment3q2arguments.txt
input_file = sc.textFile('/FileStore/tables/assignment3q2arguments.txt')
input_path = input_file.collect()
io = input_path[0].split(' ')
ipath = io[0]
opath = MOUNT_NAME + io[1][22:]
print(ipath + '\n' +opath)
```

```
display(dbutils.fs.ls(ipath))
```

```
edgesDF = spark.read.option("header","true").option("inferschema","true").csv(ipath).withColumnRenamed('id_1', 'src').withColumnRenamed('id_2', 'dst')
```

```
display(edgesDF)
edgesDF.count()
```

```
verticesDF = edgesDF.select('src').union(edgesDF.select('dst')).distinct().withColumnRenamed('src', 'id')
display(verticesDF)
verticesDF.count()
```

```
from graphframes import *
gf = GraphFrame(verticesDF,edgesDF)
```

```
query1 = gf.outDegrees.orderBy(desc('outDegree'))
query1.show(5)
```

```
query1rdd = query1.rdd.map(lambda x:(x[0], x[1]))
text = 'query1 (id, outDegree)' + '\n'
for line in query1rdd.take(5):
    text += str(line[0]) + ' ' + str(line[1])
    text += '\n'
print(text)
```

```
query2 = gf.inDegrees.orderBy(desc('inDegree'))
query2.show(5)
```

```
query2rdd = query2.rdd.map(lambda x:(x[0], x[1]))
text += 'query2 (id, inDegree)' + '\n'
for line in query2rdd.take(5):
    text += str(line[0]) + ' ' + str(line[1])
    text += '\n'
print(text)
```

```
pagerank = gf.pageRank(resetProbability=0.15, maxIter = 10)
```

```
query3 = pagerank.vertices.select("id", "pagerank").orderBy(desc('pagerank'))
query3.show(5)
```

```
query3rdd = query3.rdd.map(lambda x:(x[0], x[1]))
text += 'query3 (id, pagerank)' + '\n'
for line in query3rdd.take(5):
    text += str(line[0]) + ' ' + str(line[1])
    text += '\n'
print(text)
```

```
connComp = gf.connectedComponents()
```

```
query4 = connComp.groupBy('component').count().orderBy(desc('count'))
query4.show(5)
```

```
query4rdd = query4.rdd.map(lambda x:(x[0], x[1]))
text += 'query4 (component, count)' + '\n'
for line in query4rdd.take(5):
    text += str(line[0]) + ' ' + str(line[1])
    text += '\n'
print(text)
```

```
# since the graph is strongly connected sampling out some edges on randam to make multiple connected subgraphs.
subEdgeDF=edgesDF.sample(0.1)
subVerticesDF = subEdgeDF.select('src').union(subEdgeDF.select('dst')).distinct().withColumnRenamed('src', 'id')
subgf = GraphFrame(subVerticesDF, subEdgeDF)
```

```
subconnComp = subgf.connectedComponents()
```

```
subquery4 = subconnComp.groupBy('component').count().orderBy(desc('count'))
subquery4.show(5)
```

```
subquery4rdd = subquery4.rdd.map(lambda x:(x[0], x[1]))
text += 'subquery4 (component, count)' + '\n'
for line in subquery4rdd.take(5):
    text += str(line[0]) + ' ' + str(line[1])
    text += '\n'
print(text)
```

```
query5 = gf.triangleCount().orderBy(desc('count'))
query5.show(5)
```

```
query5rdd = query5.rdd.map(lambda x:(x[0], x[1]))
text += 'query5 (count, id)' + '\n'
for line in query5rdd.take(5):
    text += str(line[0]) + ' ' + str(line[1])
    text += '\n'
print(text)
```
