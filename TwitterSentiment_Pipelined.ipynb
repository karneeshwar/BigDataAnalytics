```
from pyspark.sql.functions import *
import urllib
ACCESS_KEY = aws_key.where(col('User name')=='karz').select('Access key ID').collect()[0]['Access key ID']
SECRET_KEY = aws_key.where(col('User name')=='karz').select('Secret access key').collect()[0]['Secret access key']
ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe="")
```

```
tweetsDF = spark.read.option("header","true").option("inferschema","true").csv("dbfs:/mnt/karz-bigdata-su22/Tweets.csv")
# or
# tweetsDF = spark.read.option("header","true").option("inferschema","true").csv("s3://karz-bigdata-su22/Tweets.csv")
```

```
display(tweetsDF)
tweetsDF.count()
```

```
tweets_filtered = tweetsDF.na.drop(subset=["text"])
display(tweets_filtered)
tweets_filtered.count()
```

```
from pyspark.ml import Pipeline
from pyspark.ml.tuning import ParamGridBuilder,CrossValidator 
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
from pyspark.ml.feature import  StopWordsRemover,Tokenizer, HashingTF, StringIndexer
```

```
train, test = tweets_filtered.randomSplit([0.80, 0.20])
```

```
tokenized = Tokenizer(inputCol = "text", outputCol = "words")
no_stopwords = StopWordsRemover(inputCol = tokenized.getOutputCol(), outputCol = "NoStopwords")
hashingTF = HashingTF(inputCol = no_stopwords.getOutputCol(), outputCol = "features")
labelledIndex = StringIndexer(inputCol = "airline_sentiment", outputCol = "label")
lr = LogisticRegression(maxIter = 10,labelCol = 'label',featuresCol = 'features')


pipeline = Pipeline(stages=[tokenized, no_stopwords, hashingTF, labelledIndex, lr])
paramGrid = ParamGridBuilder().addGrid(hashingTF.numFeatures, [10, 100, 1000, 10000]).addGrid(lr.regParam, [0.8, 0.5, 0.1, 0.05, 0.01, 0.005, 0.001]).build()


evaluator = MulticlassClassificationEvaluator(labelCol = "label", predictionCol = "prediction")
cross_validator = CrossValidator(estimator = pipeline, estimatorParamMaps = paramGrid, evaluator = evaluator, numFolds = 10)
```

```
crossvalidator_model = cross_validator.fit(train)
```

```
test_model = crossvalidator_model.transform(test)
display(test_model)
```

```
# label column is the numerical label given for the airline_sentiment, 0-negative, 1-neutral, and 2-positive
# prediction column is the predicted numerical value for the sentiments
predict = test_model.select("prediction", "label").rdd.map(lambda lp:(float(lp.prediction), float(lp.label)))
predict.collect()
```

```
metrics = MulticlassMetrics(predict)
```

```
labels_list = test_model.rdd.map(lambda lp: lp.label).distinct().collect()
```

```
text = "Overall performance metrics: \n"
text += "Accuracy            : " + str(metrics.accuracy) + "\n"
text += "Weighted Precision  : " + str(metrics.weightedPrecision) + "\n"
text += "Weighted Recall     : " + str(metrics.weightedRecall) + "\n"
text += "Weighted F1 Score   : " + str(metrics.weightedFMeasure()) + "\n"
text += "True Positive Rate  : " + str(metrics.weightedTruePositiveRate) + "\n"
text += "False Positive Rate : " + str(metrics.weightedFalsePositiveRate) + "\n"
text += "Confusion Matrix    : \n" + str(metrics.confusionMatrix()) + "\n"

text += "\n\nLabel wise performance metrics: \n"
for label in labels_list:
    if label == 0.0:
        text += "Label : " + str(label) + ", negative" + "\n"
    elif label == 2.0:
        text += "Label : " + str(label) + ", positive" + "\n" 
    else:
        text += "Label : " + str(label) + ", neutral" + "\n"
    text += "False Positive Rate :" + str(metrics.falsePositiveRate(label)) + "\n"
    text += "True Positive Rate  :" + str(metrics.truePositiveRate(label)) + "\n"
    text += "Precision           :" + str(metrics.precision(label)) + "\n"
    text += "recall              :" + str(metrics.recall(label)) + "\n"
    text += "F1 Score            :" + str(metrics.fMeasure(label, beta=1.0)) + "\n"
    
print(text)
```

```
dbutils.fs.put("/mnt/karz-bigdata-su22/TweetsOutput.txt", text)
# The text file has been written to the AWS S3 bucket and it can be accessed by URI: "s3://karz-bigdata-su22/TweetsOutput.txt"
```
