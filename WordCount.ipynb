```
from sparknlp.base import *
from sparknlp.annotator import *
from sparknlp.pretrained import PretrainedPipeline
import sparknlp
```

```
sparknlp.version()
```

```
spark = sparknlp.start()
```

```
pipeline = PretrainedPipeline('explain_document_dl', lang='en')
```

```
book = sc.textFile('/FileStore/tables/book.txt')
```

```
words = book.flatMap(lambda x: x.split(" ")).filter(lambda x: len(x)>2).collect()
```

```
import re

for i in range(len(words)):
    words[i] = re.sub('[^A-Za-z0-9]+', '', words[i])
```

```
sentence = ' '.join(words)
```

```
#symbols = ['`','~','!','@','#','$','%','^','&','*','(',')','=','+','[',']','{','}','|',':',';','/','<','>','?',"\'",'\"',',','.']

#sentence = sentence.translate( {ord(symbol): '' for symbol in symbols} )
```

```
result = pipeline.annotate(sentence)
```

```
named_entities = sc.parallelize(result['entities'])
```

```
word_count = named_entities.map(lambda x: (x,1)).reduceByKey(lambda x,y: x + y).sortBy(lambda x: -x[1]).collect()
```
